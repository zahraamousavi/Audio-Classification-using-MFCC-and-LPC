**import modules**

```python
import numpy as np
import matplotlib.pyplot as plt
import warnings
import os
from scipy.io import wavfile
import sys
sys.path.append('/gdrive/MyDrive/Digital Speech Processing/HW#03/utils')
import FeatureExtractor
import Distributions
warnings.filterwarnings('ignore')
```

**read files**

```python
base_address = '/gdrive/MyDrive/Digital Speech Processing/HW#03/'
one_names, six_names, nine_names = [], [], []
one_names_test, nine_names_test = [], []
ones, sixes, nines = [], [], []
ones_test, nines_test = [], []

# save name of all files
for file_name in os.listdir(base_address + 'data/TrainSet/1'):
  one_names.append(file_name)
for file_name in os.listdir(base_address + 'data/TrainSet/6'):
  six_names.append(file_name)
for file_name in os.listdir(base_address + 'data/TrainSet/9'):
  nine_names.append(file_name)
# test data
for file_name in os.listdir(base_address + 'data/TestSet/1'):
  one_names_test.append(file_name)
for file_name in os.listdir(base_address + 'data/TestSet/9'):
  nine_names_test.append(file_name)

# save signals of each number
for item in one_names:
  ones.append(wavfile.read(base_address+'data/TrainSet/1/'+item)[1])
for item in six_names:
  sixes.append(wavfile.read(base_address+'data/TrainSet/6/'+item)[1])
for item in nine_names:
  nines.append(wavfile.read(base_address+'data/TrainSet/9/'+item)[1])
# test data
for item in one_names_test:
  ones_test.append(wavfile.read(base_address+'data/TestSet/1/'+item)[1])
for item in nine_names_test:
  nines_test.append(wavfile.read(base_address+'data/TestSet/9/'+item)[1])
```

# Part A

calculate powers

```python
powers_one,  powers_six, powers_nine= [], [], []
powers_one_test, powers_nine_test= [], []
feature_extractor = FeatureExtractor.FeatureExtractor()
# get power of each signal
for one in ones:
  powers_one.append(feature_extractor.get_power(one))
for six in sixes:
  powers_six.append(feature_extractor.get_power(six))
for nine in nines:
  powers_nine.append(feature_extractor.get_power(nine))
for one in ones_test:
  powers_one_test.append(feature_extractor.get_power(one))
for nine in nines_test:
  powers_nine_test.append(feature_extractor.get_power(nine))

# sort powers
powers_one.sort()
powers_six.sort()
powers_nine.sort()
powers_one_test.sort()
powers_nine.sort()
# plot powers
plt.figure(figsize=(20, 4))
plt.title("power of 'one' signal")
plt.hist(powers_one, color='red')
plt.figure(figsize=(20, 4))
plt.title("power of 'six' signal")
plt.hist(powers_six, color='green')
plt.figure(figsize=(20, 4))
plt.title("power of 'nine' signal")
plt.hist(powers_nine, color='blue')
```

gaussian distribution

```python
from numpy.core.fromnumeric import size
mean_one, mean_six, mean_nine = 0, 0, 0
var_one, var_six, var_nine = 0, 0, 0
gaus_one, gaus_six, gaus_nine = [], [], []
# mean of powers
mean_one = np.mean(powers_one)
mean_six = np.mean(powers_six)
mean_nine = np.mean(powers_nine)
# standard deviation of powers
std_one = np.std(powers_one)
std_six = np.std(powers_six)
std_nine = np.std(powers_nine)
# gaussian distribution
distributions = Distributions.Distributions()
gaus_one = distributions.gaussian_distribution(powers_one, mean_one, std_one)
gaus_six = distributions.gaussian_distribution(powers_six, mean_six, std_six)
gaus_nine = distributions.gaussian_distribution(powers_nine, mean_nine, std_nine)
# plot gaussian distributions
plt.figure(figsize=(20, 4))
plt.title("gaussian distribution of 'one' signal")
plt.plot(powers_one,gaus_one, color="red")
plt.figure(figsize=(20, 4))
plt.title("gaussian distribution of 'six' signal")
plt.plot(powers_six,gaus_six, color="green")
plt.figure(figsize=(20, 4))
plt.title("gaussian distribution of 'nine' signal")
plt.plot(powers_nine,gaus_nine, color="blue")
```

decision boundary

```python
# plot distribution of power of two signal one and nine
plt.figure(figsize=(20, 4))
plt.plot(powers_one,gaus_one, color="red", label='one')
plt.plot(powers_nine,gaus_nine, color="blue", label='nine')
plt.gca().legend()
```

**set decision boundary**
<br>**First approach**
<br>set decision boundary without mathematical calculations

```python
thresh = 0.9e7
print('Threshold: 0.9e7')
counter_one, counter_nine = 0, 0
# one
for power in powers_one_test:
  if power > thresh:
    counter_one += 1
accuracy_one = counter_one / len(ones_test)
print('Accuracy for one: ', accuracy_one)
# nine
for power in powers_nine_test:
  if power < thresh:
    counter_nine += 1
accuracy_nine = counter_nine / len(nines_test)
print('Accuracy for nine: ', accuracy_nine)
```

**Second approach**
<br>estimate decision boundary by mathematical calculations

reduce powers array for get minimum distance

```python
powers_one_subset, powers_nine_subset = powers_one[150:170], powers_nine[150:170]
gaus_one_subset, gaus_nine_subset = gaus_one[150:170], gaus_nine[150:170]
plt.figure(figsize=(20, 4))
plt.scatter(powers_one_subset,gaus_one_subset, color="red", label='one')
plt.scatter(powers_nine_subset,gaus_nine_subset, color="blue", label='nine')
plt.gca().legend()
```

create distance matrix

```python
dist_mat = np.zeros((20, 20))
for i in range(len(powers_nine_subset)):
  for j in range(len(powers_one_subset)):
    dist_mat[i][j] = np.abs(powers_nine_subset[i] - powers_one_subset[j])
min_idx = np.where(dist_mat == np.min(dist_mat))
min_idx = tuple([i.item() for i in min_idx])
thresh = (powers_nine_subset[min_idx[0]] + powers_one_subset[min_idx[1]]) /2
print('decision boundary: ', thresh)
```

```python
plt.figure(figsize=(20, 4))
plt.scatter(powers_one_subset,gaus_one_subset, color="red", label='one')
plt.scatter(powers_nine_subset,gaus_nine_subset, color="blue", label='nine')
y1 = distributions.gaussian_distribution(thresh, mean_one, std_one)
y2 = distributions.gaussian_distribution(thresh, mean_nine, std_nine)
y = (y1 + y2) / 2
plt.scatter(thresh, y, color='green', label='thresh')
plt.gca().legend()
```

calculate accuracy

```python
counter_one, counter_nine = 0, 0
# one
for power in powers_one_test:
  if power > thresh:
    counter_one += 1
accuracy_one = counter_one / len(ones_test)
print('Accuracy for one: ', accuracy_one)
# nine
for power in powers_nine_test:
  if power < thresh:
    counter_nine += 1
accuracy_nine = counter_nine / len(nines_test)
print('Accuracy for nine: ', accuracy_nine)
```

**Second approach**
<br>estimate decision boundary by mathemarical calculations

reduce zcr array for get minimum distance
<br>first threshold

```python
# first threshold
zcr_ones_subset, zcr_nines_subset = zcr_ones[30:80], zcr_nines[30:80]
gaus_zcr_one_subset, gaus_zcr_nine_subset = gaus_zcr_one[30:80], gaus_zcr_nine[30:80]
plt.figure(figsize=(20, 4))
plt.scatter(zcr_ones_subset,gaus_zcr_one_subset, color="red", label='one')
plt.scatter(zcr_nines_subset,gaus_zcr_nine_subset, color="blue", label='nine')
plt.gca().legend()
```

distance matrix for first decision boundary

```python
dist_mat = np.zeros((50, 50))
for i in range(len(zcr_nines_subset)):
  for j in range(len(zcr_ones_subset)):
    dist_mat[i][j] = np.abs(zcr_nines_subset[i] - zcr_ones_subset[j])
min_idx = np.where(dist_mat == np.min(dist_mat))
min_idx = tuple([i[0] for i in min_idx])
thresh1 = (zcr_nines_subset[min_idx[0]] + zcr_ones_subset[min_idx[1]]) /2
print('first decision boundary: ', thresh1)
```

Second threshold

```python
# second threshold
zcr_ones_subset, zcr_nines_subset = zcr_ones[170:], zcr_nines[170:]
gaus_zcr_one_subset, gaus_zcr_nine_subset = gaus_zcr_one[170:], gaus_zcr_nine[170:]
plt.figure(figsize=(20, 4))
plt.scatter(zcr_ones_subset,gaus_zcr_one_subset, color="red", label='one')
plt.scatter(zcr_nines_subset,gaus_zcr_nine_subset, color="blue", label='nine')
plt.gca().legend()
```

distance matrix for second decision boundary

```python
dist_mat = np.zeros((20, 20))
for i in range(len(zcr_nines_subset)):
  for j in range(len(zcr_ones_subset)):
    dist_mat[i][j] = np.abs(zcr_nines_subset[i] - zcr_ones_subset[j])
min_idx = np.where(dist_mat == np.min(dist_mat))
min_idx = tuple([i[0] for i in min_idx])
thresh2 = (zcr_nines_subset[min_idx[0]] + zcr_ones_subset[min_idx[1]]) /2
print('second decision boundary: ', thresh2)
```

calculate accuracy

```python
counter_one, counter_nine = 0, 0
for one in zcr_ones_test:
  if one < thresh1 or one > thresh2:
    counter_one += 1
accuracy_one = counter_one / len(ones_test)
print('Accuracy of one: ', accuracy_one)
for nine in zcr_nines_test:
  if nine > thresh1 and nine < thresh2:
    counter_nine += 1
accuracy_nine = counter_nine / len(nines_test)
print('Accuracy of nine: ', accuracy_nine)
```
